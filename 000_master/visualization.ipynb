```python
import numpy as np
import pandas as pd

# Generar datos de ejemplo
np.random.seed(0)
x = np.random.rand(100)
y = x * 2 + np.random.normal(0, 0.1, 100)

# Crear un DataFrame con los datos
data = pd.DataFrame({'X': x, 'Y': y})
```

```python
import matplotlib.pyplot as plt
# Gr fica de dispersi n b sica
plt.scatter(data['X'], data['Y'])
plt.title('Gr fica de Dispersi n con Matplotlib')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import matplotlib.pyplot as plt
# Gr fica de dispersi n personalizada
plt.figure(figsize=(10, 6))
plt.scatter(data['X'], data['Y'], color='red', marker='x')
plt.title('Gr fica de Dispersi n Personalizada con Matplotlib')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
# Gr fica de dispersi n b sica con Seaborn
sns.scatterplot(x='X', y='Y', data=data)
plt.title('Gr fica de Dispersi n con Seaborn')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Gr fica de dispersi n personalizada con Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='X', y='Y', data=data, s=100, color='green', marker='o', edgecolor='black')
plt.title('Gr fica de Dispersi n Personalizada con Seaborn')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.grid(True)
plt.show()
```

```python
# Ejemplo básico de uso de NumPy
import numpy as np

# Ejemplo básico de uso de NumPy
array = np.array([1, 2, 3, 4, 5])
print(array)
```

```python
# Creación de matrices en NumPy
# Crear una matriz a partir de una lista
matrix = np.array([[1, 2, 3], [4, 5, 6]])

# Crear una matriz de ceros
zeros_matrix = np.zeros((3, 3))

# Crear una matriz de unos
ones_matrix = np.ones((2, 4))

# Crear una matriz con valores aleatorios
random_matrix = np.random.rand(3, 3)

print(matrix)
print(zeros_matrix)
print(ones_matrix)
print(random_matrix)
```

```python
# Indexación y selección de elementos en NumPy
# Acceder a un elemento específico
element = matrix[0, 1]

# Acceder a una fila completa
row = matrix[1, :]

# Acceder a una columna completa
column = matrix[:, 2]

print(element)
print(row)
print(column)
```

```python
# Operaciones matemáticas básicas en NumPy
# Suma
sum_matrix = matrix + 2

# Resta
sub_matrix = matrix - 1

# Multiplicación
mul_matrix = matrix * 2

# División
div_matrix = matrix / 2

print(sum_matrix)
print(sub_matrix)
print(mul_matrix)
print(div_matrix)
```

```python
# Funciones trigonométricas en NumPy
# Funciones trigonométricas
sin_values = np.sin(matrix)
cos_values = np.cos(matrix)

print(sin_values)
print(cos_values)
```

```python
# Evaluación de expresiones matemáticas en NumPy
# Evaluar una expresión matemática para un rango de valores de x
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.cos(x)

print(y)
```

```python
# Filtrado de datos en NumPy
# Filtrar valores mayores a un umbral
filtered = matrix[matrix > 3]

print(filtered)
```

```python
# Ordenamiento de matrices en NumPy
# Ordenar una matriz
sorted_array = np.sort(matrix, axis=0)

print(sorted_array)
```

```python
# Implementación de regresión lineal simple en NumPy
from numpy.linalg import inv

X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
y = np.dot(X, np.array([1, 2])) + 3
beta = inv(X.T.dot(X)).dot(X.T).dot(y)

print(beta)
```

```python
# Operaciones de álgebra lineal en NumPy
# Suma de matrices
sum_matrix = np.add(matrix, matrix)

# Resta de matrices
sub_matrix = np.subtract(matrix, matrix)

# Multiplicación de matrices
mul_matrix = np.dot(matrix, matrix.T)

# Transpuesta de una matriz
transposed_matrix = np.transpose(matrix)

print(sum_matrix)
print(sub_matrix)
print(mul_matrix)
print(transposed_matrix)
```

```python
# Resolución de sistemas de ecuaciones lineales en NumPy
# Resolver un sistema de ecuaciones lineales
coeff = np.array([[2, 1], [1, 3]])
constants = np.array([8, 18])
solution = np.linalg.solve(coeff, constants)

print(solution)
```

```python
# Generación de números aleatorios en NumPy
# Generar números aleatorios de una distribución uniforme
uniform_randoms = np.random.uniform(0, 1, 10)

# Generar números aleatorios de una distribución normal
normal_randoms = np.random.normal(0, 1, 10)

print(uniform_randoms)
print(normal_randoms)
```

```python
# Operaciones estadísticas básicas en NumPy
# Calcular media
mean_value = np.mean(matrix)

# Calcular mediana
median_value = np.median(matrix)

# Calcular desviación estándar
std_value = np.std(matrix)

# Calcular correlación
correlation = np.corrcoef(matrix, rowvar=False)

print(mean_value)
print(median_value)
print(std_value)
print(correlation)
```
