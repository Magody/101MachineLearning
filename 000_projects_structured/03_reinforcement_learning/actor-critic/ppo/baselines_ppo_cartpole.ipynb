```python
import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.evaluation import evaluate_policy
import time

# Crear el entorno de Gymnasium
env = gym.make('CartPole-v1')
# Crear el modelo PPO con una política de perceptrón multicapa (MLP)
model = PPO('MlpPolicy', env, verbose=1)
```

```python

# Entrenar el modelo durante 10000 pasos de tiempo
model.learn(total_timesteps=100000)

# Evaluar el rendimiento del modelo entrenado
mean_reward, std_reward = evaluate_policy(model, env, n_eval_episodes=10)

print(f"Recompensa media: {mean_reward} +/- {std_reward}")

# Cerrar el entorno
env.close()
```

```python
env = gym.make('CartPole-v1', render_mode='human')

# Visualizar el agente en acción
obs, info = env.reset()

for _ in range(1000):
    # Predecir la acción basándose en la observación actual
    action, _states = model.predict(obs)
    
    # Tomar un paso en el entorno
    obs, rewards, terminated, truncated, info = env.step(action)
    
    # Renderizar el entorno
    env.render()
    time.sleep(0.01)
    
    # Verificar si el episodio ha terminado
    if terminated or truncated:
        obs, info = env.reset()

# Cerrar el entorno
env.close()
```
